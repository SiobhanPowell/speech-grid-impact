# BAU scenario, 2030: 

```python
timer_names = {'':'Timers9pm', '_midnighttimers':'Timers12am', '_NoTimers':'TimersNone'}
maxpen_results1 = {}
maxpen_results10 = {}

fuel = 1
ev_scenario = 'BusinessAsUsual'
ev_workplace_bool = True
ev_timers = ''
wp_control = 'minpeak'
ev_scenario_date = '20211119'
for solar, wind in {2.5:2.5, 1:1, 5:5}.items():
    grid = FutureGrid(gd_short)
    max_penlevel1, max_penlevel10 = grid.find_capacity_limit_10_binarysearch(year=2030, bs_limits=None, lims_8760=limits_2030_long['limit2'].values, 
                                                           solar=solar, wind=wind, fuel=fuel,
                                                           ev_scenario=ev_scenario, ev_timers=ev_timers, 
                                                           ev_workplace_control=wp_control, ev_workplace_bool=ev_workplace_bool, 
                                                           evs_bool=True, ev_scenario_date=ev_scenario_date)

    print('fuel'+str(fuel)+'_solar'+str(solar)+'_wind'+str(wind)+'_'+ev_scenario+'_WPcontrol_'+wp_control)
    maxpen_results1['fuel'+str(fuel)+'_solar'+str(solar)+'_wind'+str(wind)+'_'+ev_scenario+'_WPcontrol_'+wp_control] = max_penlevel1
    maxpen_results10['fuel'+str(fuel)+'_solar'+str(solar)+'_wind'+str(wind)+'_'+ev_scenario+'_WPcontrol_'+wp_control] = max_penlevel10

maxpen_results2 = pd.DataFrame(maxpen_results1, index=['TimersMixed'])
maxpen_results2.to_csv('Results/max_penetration_levels_1h_bau_20211119.csv')
maxpen_results20 = pd.DataFrame(maxpen_results10, index=['TimersMixed'])
maxpen_results20.to_csv('Results/max_penetration_levels_10h_bau_20211119.csv')
```
